```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sage.all import *
```

```python
def showGraph(g, num):
    plot = g.plot(edge_labels=True)
    plot.save(f"graph{num}.png")

def displayGraph(arr, num):
    newArr = []
    for i in range(len(arr)):
        if arr[i] == 0:
            break
        newArr.append(arr[i])
    size = int(np.sqrt(len(newArr)))
    newArr = np.array(newArr)
    mat = newArr.reshape((size, size))
    a = Matrix(ZZ, mat)
    CM = CoxeterMatrix(a)
    g = CM.coxeter_graph()
    showGraph(g, num)
```

```python

train = pd.read_csv("train_data.csv")
train.head()

```

```python
test= pd.read_csv("test_data.csv")
test.head()
```

```python
Y_train = train["label"]
X_train = train.drop(labels = ["label"],axis = 1) 
X_train.head()
```

```python
X_test = test.drop(labels = ["label"],axis = 1)
y_test = test["label"]
X_test.head()
```

```python

arr = X_train.iloc[0].to_numpy()
arr1 = X_train.iloc[1].to_numpy()
displayGraph(arr, 0)
displayGraph(arr1, 1)
```

```python
# Normalize the data
X_train = X_train / 7
test = test / 7
print("x_train shape: ",X_train.shape)
print("test shape: ",test.shape)
```

```python
# Reshape
X_train = X_train.values.reshape(-1,9,9,1)
X_test = X_test.values.reshape(-1,9,9,1)
print("x_train shape: ",X_train.shape)
print("x_test shape: ",X_test.shape)

```

```python


# Label Encoding 
from tensorflow.keras.utils import to_categorical # convert to one-hot-encoding
Y_train = to_categorical(Y_train, num_classes = 4)
y_test = to_categorical(y_test, num_classes = 4)
```

```python


# Split the train and the validation set for the fitting
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.1, random_state=2)
print("x_train shape",X_train.shape)
print("x_test shape",X_val.shape)
print("y_train shape",Y_train.shape)
print("y_test shape",Y_val.shape)

```

```python
from sklearn.metrics import confusion_matrix
import itertools

from tensorflow.keras.utils import to_categorical # convert to one-hot-encoding
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from keras.optimizers import RMSprop,Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau
model = Sequential()
model.add(Conv2D(8, kernel_size=(3,3), padding='Same', activation='relu', input_shape=(9,9,1)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(64, activation='relu'))  # fewer neurons
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))  # 4 output classes
```

```python
optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
```

```python
# Compile the model
model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(X_train, Y_train, epochs=10, batch_size=64, validation_data=(X_val, Y_val))
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_accuracy:.4f}")
```
